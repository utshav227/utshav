```python
ug['START_DATE*']=ug['START_DATE*'].astype('datetime64[ns]')
ug['END_DATE*']=ug['END_DATE*'].astype('datetime64[ns]')
```

```python
ug['Time'] = ug['END_DATE*']-ug['START_DATE*'] #find time required by substracting the two values
ug.head()
```

```python
ug['Time'] = ug['Time'].dt.total_seconds()/60.0
ug['Time'] = ug['Time'].astype('float')
ug.head()
```

```python
ug['KM']= ug['MILES*']*1.609 #converted miles to km for better understanding
```

```python
ug.head()
```

```python
ug.drop(['MILES*'],axis=1) #dropped miles since km is present
```

```python
ug['speed']=(ug['KM']/ug['Time'])*60 #time:min & speed:kmph
```

```python
#HOUR DAY TRANSFORMATION
```

```python
import datetime
import calendar
```

```python
ug['START_DATE*'] = pd.to_datetime(ug['START_DATE*'], format="%m/%d/%Y %H:%M") 
ug['END_DATE*'] = pd.to_datetime(ug['END_DATE*'], format="%m/%d/%Y %H:%M")
```

```python
hour=[]  #empty list
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in ug['START_DATE*']:
    hour.append(x.hour) #adding/appending the values to above empty list
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
ug['HOUR']=hour #creatig columns
ug['DAY']=day
ug['DAY_OF_WEEK']=dayofweek
ug['MONTH']=month
ug['WEEKDAY']=weekday
```

```python
ug.head() #updated dataset
```

```python
ug['HOUR'].value_counts().plot(kind='bar',figsize=(10,5),color='red') #trips per hour
```

```python

ug['WEEKDAY'].value_counts().plot(kind='bar',figsize=(10,5),color='black') #trips per day of week
```

```python
ug['DAY'].value_counts().plot(kind='bar',figsize=(10,5),color='blue') #trips per day of the month
```

```python
ug['MONTH'].value_counts().plot(kind='bar',figsize=(10,5),color='green') #trips in month, month 12 has highest number of trips as depicted in the figure
```
